## 1. Entendimento do Negócio

### 1.1 Qual problema de negócio está sendo resolvido?

O problema central é a **incapacidade da empresa de antecipar a insatisfação do cliente**. Hoje, o NPS só é coletado após o encerramento da jornada de compra, ou seja, quando o "estrago" já está feito. A empresa opera de forma **reativa**: descobre que o cliente ficou insatisfeito apenas depois que ele já teve a experiência negativa, já formou sua opinião e, potencialmente, já compartilhou essa opinião com outros consumidores.

O desafio propõe inverter essa lógica: usar dados operacionais que a empresa **já possui** (informações de pedido, logística e atendimento) para **prever** qual será o NPS do cliente antes da avaliação ocorrer. Isso transforma a gestão de experiência de reativa para **proativa e preventiva**.

Em termos práticos, estamos resolvendo três problemas interligados:

1. **Problema de diagnóstico:** Quais fatores operacionais realmente determinam se um cliente será promotor ou detrator?
2. **Problema de priorização:** Onde a empresa deve concentrar recursos para obter o maior impacto na satisfação?
3. **Problema de antecipação:** É possível identificar um cliente em risco de insatisfação durante a jornada de compra, antes da pesquisa NPS?

### 1.2 Por que o NPS é importante para um e-commerce?

O NPS é particularmente crítico para e-commerce por razões que se retroalimentam:

**Custo de aquisição vs. retenção:** No e-commerce brasileiro, o custo de adquirir um novo cliente (CAC) pode ser 5 a 7 vezes maior do que reter um existente. Um detrator não apenas deixa de comprar, ele eleva o CAC da empresa, pois cada cliente perdido precisa ser substituído por um novo, adquirido via marketing.

**Transparência radical do digital:** Diferente do varejo físico, no e-commerce a insatisfação é pública e permanente. Avaliações negativas ficam visíveis em marketplaces, redes sociais e sites de reclamação (como Reclame Aqui), impactando diretamente a decisão de compra de futuros clientes.

**Margens comprimidas:** O e-commerce brasileiro opera com margens apertadas, especialmente considerando frete, logística reversa e custos de atendimento. Um NPS baixo gera custos ocultos (mais contatos de atendimento, mais devoluções, mais reclamações públicas), corroendo a rentabilidade.

### 1.3 Quais áreas poderiam se beneficiar desses insights?

| Área | Como se beneficia | Exemplo de ação |
|------|-------------------|-----------------|
| **Logística** | Identificar que atraso na entrega é o principal destruidor de NPS | Priorizar entregas com alto risco de atraso; renegociar SLAs com transportadoras |
| **Atendimento ao Cliente** | Saber que múltiplos contatos amplificam insatisfação | Resolver no primeiro contato (FCR); acionar proativamente clientes com pedidos problemáticos |
| **Pricing / Comercial** | Entender a relação entre frete, desconto e percepção de valor | Calibrar frete grátis como investimento em NPS, não como custo |
| **Produto / UX** | Identificar perfis de cliente mais sensíveis | Adaptar comunicação e expectativas por segmento |
| **Estratégia / C-Level** | Ter um modelo preditivo de NPS como KPI antecedente | Dashboards com NPS previsto em tempo real para tomada de decisão |
| **Marketing** | Segmentar promotores para programas de indicação | Ativar promotores como embaixadores; conter detratores antes que publiquem |

### 1.4 Reflexão: Impacto do NPS no Negócio

#### NPS e Recompra

A relação entre NPS e recompra é direta e mensurável. Promotores (NPS 9-10) têm probabilidade de recompra significativamente maior que detratores (NPS 0-6). Essa relação opera por dois mecanismos:

- **Confiança construída:** A confiança é o atalho que faz o cliente comprar de novo sem precisar buscar outras empresas.
- **Custo de troca:** Clientes satisfeitos acumulam "capital relacional" (cadastro, histórico, cupons), tornando a troca para um concorrente um custo percebido.

Em nossos dados, a variável `repeat_purchase_30d` captura exatamente isso, e sua alta correlação positiva com NPS (r ≈ 0.57) confirma essa dinâmica.

#### NPS e Boca a Boca

O boca a boca é o mecanismo mais poderoso (e gratuito) de aquisição, e o NPS é seu melhor preditor. No e-commerce:

- **Promotores** recomendam ativamente em redes sociais, grupos de WhatsApp e avaliações em marketplaces.
- **Detratores** fazem o inverso com intensidade ainda maior: pesquisas de comportamento do consumidor indicam que clientes insatisfeitos compartilham experiências negativas com 2 a 3 vezes mais pessoas do que clientes satisfeitos compartilham experiências positivas.

A assimetria é crítica: **o impacto negativo do detrator é proporcionalmente maior que o impacto positivo do promoto**r.

#### NPS e Market Share

No e-commerce, market share é uma batalha de percepção. O NPS impacta market share por três vias:

1. **Retenção → base estável:** NPS alto retém clientes, garantindo receita recorrente.
2. **Boca a boca → aquisição orgânica:** Promotores trazem novos clientes sem custo de marketing.
3. **Reputação digital → conversão:** Avaliações positivas aumentam a taxa de conversão de visitantes em compradores.

Empresas com NPS consistentemente alto crescem cerca de 2x mais rápido que seus concorrentes do mesmo setor, pois o ciclo virtuoso (retenção + aquisição orgânica + melhor conversão) se retroalimenta.

## 2. Definição da Target (Variável Alvo)

### 2.1 Qual variável representa a satisfação do cliente?

A variável alvo é o `nps_score`, a nota de satisfação do cliente na escala NPS, variando de 0 a 10, coletada após a experiência de compra.

### 2.2 Por que ela foi escolhida?

O `nps_score` é a escolha natural por múltiplas razões:

1. **É a métrica padrão de mercado** para satisfação do cliente, permitindo benchmarking com concorrentes e com o setor.
2. **Captura a experiência completa**, ao ser coletada pós-compra, reflete o efeito acumulado de todas as etapas da jornada (pedido, pagamento, entrega, eventual atendimento).
3. **É acionável**, a escala 0-10 permite categorização direta em Detratores (0-6), Neutros (7-8) e Promotores (9-10), cada grupo demandando ações diferentes.
4. **Tem variabilidade suficiente**, na base, o NPS varia de 0 a 10 com desvio padrão de 2.51, o que significa que há "sinal" suficiente para um modelo aprender padrões.

### 2.3 Em que momento da jornada essa informação é coletada?

O NPS é coletado **após o encerramento da jornada de compra**, ou seja, depois que o pedido foi entregue (ou não), eventuais problemas foram (ou não) resolvidos, e o cliente já formou sua opinião sobre a experiência completa.

Esse timing é crucial para o problema de negócio: como a coleta é ex-post, a empresa **não consegue agir preventivamente**. O modelo preditivo proposto visa justamente preencher essa lacuna, prever o NPS usando dados disponíveis durante a jornada, antes da pesquisa.

### 2.4 Existe algum risco de usar essa variável de forma inadequada?

Sim, há riscos relevantes que precisam ser considerados:

**Risco 1 — Viés de resposta:** Nem todo cliente responde à pesquisa NPS. Clientes muito satisfeitos ou muito insatisfeitos tendem a responder mais, enquanto neutros respondem menos. A base pode não representar fielmente a distribuição real de satisfação.

**Risco 2 — Possível data leakage em variáveis correlatas:** A base contém duas variáveis com alta correlação positiva com o NPS que merecem análise cuidadosa quanto ao risco de **data leakage** (vazamento de informação temporal):

---
 ```repeat_purchase_30```
* **Correlação:** 0.57
* **Risco:** 🔴 ALTO
* **Justificativa:** A recompra em 30 dias é um evento que acontece **depois** da pesquisa de NPS, o que invalida a previsão em tempo real.

---
```csat_internal_score```
* **Correlação:** 0.56
* **Risco:** 🟠 MODERADO
* **Justificativa:** É uma métrica da empresa (eficiência). Só é útil se o cálculo for fechado **antes** de disparar o NPS; caso contrário, é apenas um eco do resultado.
---


> **Decisão do projeto:** Optamos por **manter ambas as variáveis** na análise e no modelo, pelos seguintes motivos:
>
> 1. **Transparência analítica:** Remover variáveis sem confirmação do timing real seria uma decisão baseada em suposição. Preferimos documentar o risco e deixar a decisão final para a equipe de negócio.
> 2. **Valor pedagógico:** Manter as variáveis permite demonstrar seu impacto no modelo e discutir explicitamente o que aconteceria se fossem removidas.
> 3. **Cenário de produção:** Em deploy real, a empresa **deve** confirmar a ordem temporal de coleta. Se `repeat_purchase_30d` for confirmada como posterior ao NPS, ela deve ser removida do modelo. Se `csat_internal_score` for computado a partir de dados operacionais já disponíveis, pode (e deve) ser mantida.
>
> Essa discussão será retomada no Notebook 03 (Modelo Preditivo), onde analisaremos o impacto de cada variável no desempenho do modelo.

**Risco 3 — Tratar NPS como variável contínua vs. ordinal:** NPS é uma escala de 0 a 10, mas não necessariamente a "distância" entre 3 e 4 é a mesma que entre 8 e 9 em termos de comportamento do cliente. Um modelo de regressão assume linearidade; um de classificação (detrator/não-detrator) pode capturar melhor a não-linearidade da escala.

**Risco 4 — NPS como métrica única de sucesso:** O NPS é um indicador importante, mas não deve ser o único. Otimizar exclusivamente para NPS pode levar a decisões que melhoram a nota mas não geram valor real (ex: dar desconto excessivo para "comprar" satisfação).
